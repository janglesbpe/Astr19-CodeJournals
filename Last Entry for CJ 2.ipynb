```python
#Imports
import numpy as np
import matplotlib.pyplot as mlp
```

```python
#number generate
random_numbers = np.random.normal(loc=0, scale=1, size=1000)
```

```python
#Plot Graph
mlp.hist(random_numbers, bins=100, edgecolor='black')
mlp.xlabel("Numbers")
mlp.ylabel("Frequency")
mlp.title("Histogram of Random Numbers")
mlp.show()
```
