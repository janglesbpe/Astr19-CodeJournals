```python
#Imports
import numpy as np
import matplotlib.pyplot as mlp
```

```python
#Py Random Number Pull
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
#Plot
mlp.hist(random_numbers, bins=100, edgecolor='black')
mlp.xlabel("Number")
mlp.ylabel("Frequency")
mlp.title("Histogram of Random Numbers")
mlp.show()
```
