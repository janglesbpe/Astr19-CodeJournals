```python
#Imports 
```

```python
import numpy as np 
import matplotlib.pyplot as mlp
```

```python
#Create array
```

```python
x= np.linspace(0,1,100)
```

```python
#Define function
```

```python
def exp_function(x):
    return np.exp(x)
y = exp_function(x)
```

```python
#Plot x vs. y
```

```python
mlp.plot(x,y)
mlp.xlabel("Time [milliseconds]")
mlp.ylabel("Awesomeness")
mlp.title("Exponential Growth of Awesomeness over Time")
mlp.show()
```
